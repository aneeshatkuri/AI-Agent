```python
from langchain_pinecone import PineconeVectorStore
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.llms import OpenAI
from langchain.chains import RetrievalQA
from langchain_openai import ChatOpenAI

# Initialize embeddings model
embeddings_model = OpenAIEmbeddings(model="text-embedding-ada-002")

# Initialize Pinecone vector store
vectorstore = PineconeVectorStore(index, embeddings_model, "text")

# Initialize OpenAI LLM
llm = ChatOpenAI(temperature=0.7, model="gpt-4o", max_tokens=500, verbose=True, openai_api_key="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")

# Create RetrievalQA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vectorstore.as_retriever(),
    return_source_documents=True
)
```

```python
import re

def clean_text(text):
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)
    # Remove extra newlines
    text = re.sub(r'\n+', '\n', text)
    return text.strip()

def generate_summary(query):
    result = qa_chain({"query": query})
    
    summary = result['result']
    
    sources = []
    for doc in result.get('source_documents', []):
        if hasattr(doc, 'page_content'):
            clean_content = clean_text(doc.page_content)
            sources.append(clean_content[:500] + "..." if len(clean_content) > 500 else clean_content)
        else:
            sources.append("Source content not available")
    
    return summary, sources

# Example usage
query = "Comparing Aneesh and Pavan Kumar who do you think has the better experience in MLE/MLOps role and why?"
summary, sources = generate_summary(query)

print("Generated Summary:")
print(summary)
print("\nSources:")
for source in sources:
    print(source)
```
